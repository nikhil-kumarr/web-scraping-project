```python
jobrole1=[]
company1=[]
loc1=[]
location1=[]
openings1=[]
qualification1=[]
experience1=[]
salary1=[]
for i in range(1,121):
    URL = ' https://apna.co/jobs/female-jobs-in-mumbai?city_id=1&gender=f&fall_back_type=location&page={}'. format(i)
    page = requests.get(URL)
    htmlCode = page.text
    soup = BeautifulSoup(htmlCode)
    
    roles = soup.find_all('h3', attrs={'class' : 'styles__JobTitle-sc-1eqgvmq-5 jWtSuf'})
    for r in roles:
        jobrole1.append(r.text)
        
    company_names = soup.find_all('p', attrs={'class' : 'styles__JobOrganizationName-sc-1eqgvmq-6 oixmW'})
    for c in company_names:
        company1.append(c.text)
        
    salarys = soup.find_all('div', attrs={'class' : 'styles__SalaryText-sc-1eqgvmq-3 kMSQpi'})
    for s in salarys:
        salary1.append(s.text)  
    minsalary=[]
    maxsalary=[]
    for i in salary1:
        e=i.split('-')
        minsalary.append(e[0])
        maxsalary.append(e[-1])
    
    eligibilitys = soup.find_all('div', attrs={'class' : 'styles__TagsContainer-sc-1eqgvmq-7 hdJXNf'})
    for e in eligibilitys:
        loc1.append(e.text)
    '''location1=loc1[0::4]
    openings1=loc1[1::4]
    qualification1=loc1[2::4]
    experience1=loc1[3::4]'''
    
data={'jobRole':jobrole1,'Company':company1,'Location':loc1,'Minsalary':minsalary,'Maxsalary':maxsalary}
df=pd.DataFrame(data)
df.head()
```
